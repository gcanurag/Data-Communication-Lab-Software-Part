```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def f(t, y):
    return -4 * y + 2 * np.cos(2 * t)

# Initial condition and time span
y0 = 2
t_span = (0, 15)
t_eval = np.linspace(0, 15, 1000)  # Time points for evaluation

# Numerical solution using scipy.integrate.solve_ivp
sol = solve_ivp(f, t_span, [y0], t_eval=t_eval)

# Analytical solution (you can implement this part)
# ...

# Plotting
plt.plot(sol.t, sol.y[0], label='Numerical Solution')
# plt.plot(t_eval, analytical_solution, label='Analytical Solution')  # Uncomment and add analytical solution

plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.title('Solution of dy/dt + 4y(t) = 2cos(2t)')
plt.legend()
plt.grid(True)
plt.show()
```
