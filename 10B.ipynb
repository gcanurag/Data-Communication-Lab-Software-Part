```python
import numpy as np  
import matplotlib.pyplot as plt  

# Parameters  
frequency = 5  # Frequency in Hz  
amplitude = 9  # Amplitude in Volts  
Fs = 100       # Sampling frequency in Hz  
T = 1/Fs       # Sampling period  
t = np.arange(0, 3, T)  # Time vector from 0 to 3 seconds  

# Generate the square wave  
g_t = amplitude * np.sign(np.sin(2 * np.pi * frequency * t))  

# Perform Fourier Transform  
G_f = np.fft.fft(g_t)  

# Frequency vector for plotting  
frequencies = np.fft.fftfreq(len(t), T)  

# Perform Inverse Fourier Transform  
g_t_reconstructed = np.fft.ifft(G_f)  

# Plotting  
plt.figure(figsize=(12, 10))  

# Original Square Wave  
plt.subplot(3, 1, 1)  
plt.plot(t, g_t, label='Original Square Wave', color='blue')  
plt.title('Original Square Wave g(t)')  
plt.xlabel('Time (seconds)')  
plt.ylabel('Amplitude (Volts)')  
plt.grid()  
plt.legend()  

# Fourier Transform  
plt.subplot(3, 1, 2)  
plt.plot(frequencies, np.abs(G_f), label='Fourier Transform', color='orange')  
plt.title('Fourier Transform of the Square Wave')  
plt.xlabel('Frequency (Hz)')  
plt.ylabel('Magnitude')  
plt.grid()  
plt.legend()  

# Reconstructed Signal from IFFT  
plt.subplot(3, 1, 3)  
plt.plot(t, np.real(g_t_reconstructed), label='Reconstructed Signal', color='green')  
plt.title('Reconstructed Signal from Inverse Fourier Transform')  
plt.xlabel('Time (seconds)')  
plt.ylabel('Amplitude (Volts)')  
plt.grid()  
plt.legend()  

# Adjust layout  
plt.tight_layout()  
plt.suptitle('Square Wave: Fourier Transform and Inverse Fourier Transform', fontsize=16)  
plt.subplots_adjust(top=0.93)  # Adjust title position  
plt.show()
```
