```python
import numpy as np  
import matplotlib.pyplot as plt  

# Parameters  
frequency = 5  # Frequency in Hz  
amplitude = 9  # Amplitude in Volts  
Fs = 100       # Sampling frequency in Hz  
T = 1/Fs       # Sampling period  
t = np.arange(0, 3, T)  # Time vector from 0 to 3 seconds  

# Generate the cosine wave  
g_t = amplitude * np.cos(2 * np.pi * frequency * t)  

# Perform Fourier Transform  
G_f = np.fft.fft(g_t)  

# Perform Inverse Fourier Transform  
g_t_reconstructed = np.fft.ifft(G_f)  

# Plotting  
plt.figure(figsize=(12, 8))  

# Original Cosine Wave  
plt.subplot(2, 1, 1)  
plt.plot(t, g_t, label='Original Signal')  
plt.title('Original Cosine Wave g(t)')  
plt.xlabel('Time (seconds)')  
plt.ylabel('Amplitude (Volts)')  
plt.grid()  
plt.legend()  

# Reconstructed Signal from IFFT  
plt.subplot(2, 1, 2)  
plt.plot(t, np.real(g_t_reconstructed), label='Reconstructed Signal', color='orange')  
plt.title('Reconstructed Signal from Inverse Fourier Transform')  
plt.xlabel('Time (seconds)')  
plt.ylabel('Amplitude (Volts)')  
plt.grid()  
plt.legend()  

# Adjust layout  
plt.tight_layout()  
plt.suptitle('Fourier Transform and Inverse Fourier Transform', fontsize=16)  
plt.subplots_adjust(top=0.9)  # Adjust title position  
plt.show()
```
