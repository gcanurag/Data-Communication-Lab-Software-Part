```python
import numpy as np  
import matplotlib.pyplot as plt  

# Define the range for k  
k1 = np.arange(10, 21)  # For the first signal  
k2 = np.arange(0, 51)   # For the second signal  

# Define the signals  
f1 = -0.92 * np.sin((3 * np.pi / 4) * k1)  
f2 = -0.93 * np.exp(1j * (np.pi * k2 / 350))  

# Create the figure and axis  
plt.figure(figsize=(12, 6))  

# Plot the first signal  
plt.subplot(2, 1, 1)  
plt.stem(k1, f1, basefmt=" ")  # Removed use_line_collection
plt.title('Discrete Time Signal f[k] = -0.92 sin((3/4)k)')  
plt.xlabel('k (integer values)')  
plt.ylabel('f[k]')  
plt.grid(True)  

# Plot the second signal  
plt.subplot(2, 1, 2)  
plt.stem(k2, f2.real, basefmt=" ")  # Plotting the real part
plt.title('Discrete Time Signal f[k] = -0.93 * exp(j * (k/350))')  
plt.xlabel('k (integer values)')  
plt.ylabel('f[k]')  
plt.grid(True)  

# Show the plots  
plt.tight_layout()  
plt.show()
```
