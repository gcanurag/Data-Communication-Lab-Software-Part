```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time ranges
t1 = np.linspace(-5, 5, 1000)
t2 = np.linspace(-10, 10, 1000)

# Define the signals
x1 = 5 * np.sin(2 * np.pi * t1) * np.cos(np.pi * t1 - 8)
x2 = 5 * np.exp(-0.2 * t2) * np.sin(2 * np.pi * t2)

# Plot the first signal
plt.figure(figsize=(10, 4))
plt.plot(t1, x1, label=r'$x(t) = 5 \sin(2 \pi t) \cos(\pi t - 8)$')
plt.title(r'Continuous Time Signal $x(t) = 5 \sin(2 \pi t) \cos(\pi t - 8)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

# Plot the second signal
plt.figure(figsize=(10, 4))
plt.plot(t2, x2, label=r'$x(t) = 5e^{-0.2t} \sin(2 \pi t)$')
plt.title(r'Continuous Time Signal $x(t) = 5e^{-0.2t} \sin(2 \pi t)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
