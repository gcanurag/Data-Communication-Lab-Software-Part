```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
frequency = 1  # Hz
amplitude = 5  # Volts
dc_bias = -3  # Volts
time_interval = np.linspace(-3, 3, 1000)  # Time interval from -3 to 3 seconds

# Generate the square wave
square_wave = amplitude * np.sign(np.sin(2 * np.pi * frequency * time_interval)) + dc_bias

# Plot the square wave
plt.figure(figsize=(10, 4))
plt.plot(time_interval, square_wave)
plt.title('Square Wave g(t)')
plt.xlabel('Time (seconds)')
plt.ylabel('Voltage (Volts)')
plt.grid(True)
plt.ylim(-8, 8)
plt.show()
```
